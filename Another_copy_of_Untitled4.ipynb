```python
_, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, label in zip(axes, df.images, df.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title("Training: %i" % label)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits # Import the load_digits function

# Load the digits dataset
df = load_digits()

# Splitting data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(df.data, df.target, test_size=0.2, random_state=42) # Assuming 20% for testing and random_state for reproducibility

# Now you can proceed with fitting the model
rf.fit(x_train, y_train)
```

```python
# @title Predict Test Data
```

```python
y_pred = rf.predict(x_test)
```

```python
y_pred
```

```python
# @title Model Accurancy
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
