```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
known_data = pd.DataFrame({"y":[2,3,4,5,6,12.5],"x":[1,1.5,2,2.5,3,6.25]})
```

```python
X=known_data['x'].values.reshape(-1,1)
Y=known_data['y'].values
print(X)
print(Y)
```

```python
model=LinearRegression()
model.fit(X,Y)
```

```python
predicted_A=model.predict([[22],[11]])
```

```python
print(predicted_A)
```
